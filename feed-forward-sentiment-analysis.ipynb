```python
from collections import Counter
import numpy as np
```

It is quite interesting that words such as "the" , "." , "and" , "of" and so on **appear in both negative and positive reviews** and they are the most common ones.We can conclude that it seems redundant to count the and store the words in a positive or negative review.
> What we actually need to find are words that are more unique to positive sentences or vice versa.

**To achieve this task we simple find the ratio of the most common positive words to the most common negative negative ones.** we would store the ratios in counter object and use that object for analysis further down the road. 

```python
ratios = Counter()
# a word is "common" if they have been used at least 50 times. This way we can remove certain inconsequential words.
for word,counter in list(total.most_common()):
    if(counter>50):
        #The +1 in the denominator is to prevent dividing by zeros; 
        #ie, when word is found only in posive review,in that case ratios [word] = positive[word]
        
        ratios [word] = positive[word] / float(negative[word]+1)
```

Examine the ratios you've calculated for a few words:

```python
print("Positive to negative ratio for 'the' = {}".format(ratios["the"]))
print("Positive to negative ratio for '.' = {}".format(ratios["."]))
print("Positive to negative ratio for 'excellent' = {}".format(ratios["excellent"]))
print("Positive to negative ratio for 'weak' = {}".format(ratios["bad"]))
```

We can reac the following conclusions

* Neutral words such as "." or "the" which are common to both positive and negative review have a ratio close to one.
* words that have been used in positive reviews more often have a ratio greater than 'one' . The more a word is leaning towards a positive value , the larger its ratio is going to be. 
* words that have been used in negative reviews more often have a ratio close to 'zero' . The more a word is leaning towards a negative value , closer its ratio is going to be to zero. 

At this point , when a word is common in both reviews, it has a ratio close to one and also comparing ratios that are larger than one and ratios that are closer to zero for classification is not that useful . We would ** convert the ratios to its logarithmic value** so that most common words correspond with 'zero' , positive ratios are positive numbers and negative ratios are negative numbers.

```python
for word in (ratios):
    ratios[word] = np.log(ratios[word])
```

# Getting inputs and outputs ready

In the following block we create a `Set` that contains all of our words (ie , our vocabulary)

```python
vocabulary = set()
for i in range(len(reviews)):
    for word in reviews[i].split(' '):
        vocabulary.add(word)
```

We initialize an array with values of zero as input layer.

```python
input_layer = np.zeros((1,len(vocabulary)))
```

We meed to store the index of each word since `input_layer` is supposed to have one entry point for each word. We would store the index of each word in a `lookup table`.

```python
vocabulary_lookup_table = {}
for i,word in enumerate(vocabulary):
    vocabulary_lookup_table[word] = i
    
# display the lookup table
vocabulary_lookup_table
```

`set_input_layer` stores the number of times a word occurs in the review you pass to function and it will store it in the **corresponding index** of `input_layer` array. 

```python
def set_input_layer(review):
    """ Modify the global layer_0 to represent the vector form of review.
    The element at a given index of layer_0 should represent
    how many times the given word occurs in the review.
    Args:
        review(string) - the string of the review
    Returns:
        None
    """
    global input_layer
    #resetting previous state 
    input_layer *= 0
    for word in review.split(" "):
        input_layer[0][vocabulary_lookup_table[word]] += 1
```

```python
import time
import sys
import numpy as np
class NeuralNetwork:
    def __init__(self, reviews,labels,min_count=20,polarity_cutoff=0.8,learning_rate=0.01 ,hidden_nodes = 20):
        #To get consistent and reproducable random values, we would assign a seed 
        np.random.seed(1)
        # Pre processing and denoising
        self.data_processing(reviews, labels, polarity_cutoff, min_count)
        
        #build the network
        self.build_network(len(self.review_vocabulary),hidden_nodes, 1, learning_rate)

    ## New for Project 6: added min_count and polarity_cutoff parameters
    def data_processing(self, reviews, labels, polarity_cutoff, min_count):
        
        #-----------------------------------------------------------
        #In the following block, we will count the total number of 
        #words in positive reviews , negative reviews and total reviews and 
        #store them in the counter objects we initialized.
        
        positive = Counter()
        negative = Counter()
        total = Counter()

        for i in range(len(reviews)):
            if(labels[i] == 'positive'):
                for word in reviews[i].split(" "):
                    positive[word] += 1
                    total[word] += 1
            else:
                for word in reviews[i].split(" "):
                    negative[word] += 1
                    total[word] += 1
        #positive to negative ratio
        #-----------------------------------------------------------
        ratios = Counter()

        for word,counter in list(total.most_common()):
            if(counter >= 50):
                pos_neg_ratio = positive[word] / float(negative[word]+1)
                ratios[word] = pos_neg_ratio

        for word,ratio in ratios.most_common():
            if(ratio > 1):
                ratios[word] = np.log(ratio)
            else:
                ratios[word] = -np.log((1 / (ratio + 0.01)))
        #-----------------------------------------------------------
        # Creating review vocabulary
        review_vocabulary = set()
        for review in reviews:
            for word in review.split(" "):
                # Rules for words to be added to vocabular ( Denoising )y:
                # 1) Words have to at least accur min_count : making sure a word is used 
                # 2) Ratio has to be larger than polarity_cutoff value : reducing noise by removing most
                #recurred words such as 'the'
                if(total[word] > min_count):
                    if(word in ratios.keys()):
                        if((ratios[word] >= polarity_cutoff) or (ratios[word] <= -polarity_cutoff)):
                            review_vocabulary.add(word)
                    else:
                        review_vocabulary.add(word)

        self.review_vocabulary = list(review_vocabulary)
        self.review_vocabulary_size = len(self.review_vocabulary)
        self.review_vocabulary_lookup_table = {}
        for i, word in enumerate(self.review_vocabulary):
            self.review_vocabulary_lookup_table[word] = i
        #-----------------------------------------------------------
        # Creating label vocabulary   
            
        label_vocabulary = set()
        for label in labels:
            label_vocabulary.add(label)
        
        self.label_vocabulary = list(label_vocabulary)
        self.label_vocabulary_size = len(self.label_vocabulary)
        self.label_vocabulary_lookup_table = {}
        for i, label in enumerate(self.label_vocabulary):
            self.label_vocabulary_lookup_table[label] = i

    def build_network(self, input_nodes, hidden_nodes, output_nodes, learning_rate):
        self.input_nodes = input_nodes
        self.hidden_nodes = hidden_nodes
        self.output_nodes = output_nodes
        self.learning_rate = learning_rate

        # These are the weights between the input layer and the hidden layer. we initialize them as zero
        self.weights_input_hidden = np.zeros((self.input_nodes,self.hidden_nodes))

        # weights between the hidden and output layer.we give these a random value since it is from hidden layer to output. 
        self.weights_hidden_output = np.random.normal(0.0, self.output_nodes**-0.5, 
                                                (self.hidden_nodes, self.output_nodes))
        
        # initiazlize the input_layer as a two-dimensional matrix of 1 x hidden_nodes with zeros. 
        self.input_layer = np.zeros((1,hidden_nodes))
    
    
    def label_binarizer(self,label):
        if(label == 'positive'):
            return 1
        else:
            return 0
        
    def sigmoid(self,x):
        return 1 / (1 + np.exp(-x))
    
    def sigmoid_derivative(self,output):
        return output * (1 - output)
    
   
    def train(self, raw_training_reviews, training_labels):
        correct_predictions = 0
        start_time = time.time()
        #-----------------------------------------------------------
        # we want to access non zero indices so we put them in training_reviews list,
        training_reviews = list()
        for review in raw_training_reviews:
            indices = set()
            for word in review.split(" "):
                if(word in self.review_vocabulary_lookup_table.keys()):
                    indices.add(self.review_vocabulary_lookup_table[word])
            training_reviews.append(list(indices))


        # Here is to confirm that training_reviews and training_labels
        # have the same number of records
        assert(len(training_reviews) == len(training_labels))
        
   
        for i in range(len(training_reviews)):
            
            # Get the next review and its correct label
            review = training_reviews[i]
            label = training_labels[i]
            #----------------
            #Forward Pass
            #reset previous state
            self.input_layer *= 0
            for index in review:
                self.input_layer += self.weights_input_hidden[index]

           
            output_layer = self.sigmoid(self.input_layer.dot(self.weights_hidden_output))            
           
            #----------------
            #Backward pass
            output_layer_error = output_layer - self.label_binarizer(label) 
            output_layer_delta = output_layer_error * self.sigmoid_derivative(output_layer)

            input_layer_error = output_layer_delta.dot(self.weights_hidden_output.T) 
            input_layer_delta = input_layer_error 
            

            self.weights_hidden_output -= self.input_layer.T.dot(output_layer_delta) * self.learning_rate 
            

            for index in review:
                self.weights_input_hidden[index] -= input_layer_delta[0] * self.learning_rate 
                
            #----------------
            # record keeping
            if(output_layer >= 0.5 and label == 'positive'):
                correct_predictions += 1
            elif(output_layer < 0.5 and label == 'negative'):
                correct_predictions += 1
            total_time = float(time.time() - start_time)
            reviews_per_second = i / total_time if total_time > 0 else 0
            
            sys.stdout.write("\rCurrent Progress:" + str(100 * i/float(len(training_reviews)))[:4] \
                             + "% Processing Speed(reviews/sec):" + str(reviews_per_second)[0:5] \
                             + " Correct Predictions:" + str(correct_predictions) + " Trained:" + str(i+1) \
                             + " Training Accuracy:" + str(correct_predictions * 100 / float(i+1))[:4] + "%")
            if(i % 2500 == 0):
                print("")
    
    def test(self, testing_reviews, testing_labels):
              
        correct_predictions = 0

        start_time = time.time()
        for i in range(len(testing_reviews)):
            prediction = self.run(testing_reviews[i])
            if(prediction == testing_labels[i]):
                correct_predictions += 1
            
            total_time = float(time.time() - start_time)
            reviews_per_second = i / total_time if total_time > 0 else 0
            
            sys.stdout.write("\rCurrent Progress:" + str(100 * i/float(len(testing_reviews)))[:2] \
                             + "% Processing Speed(reviews/sec):" + str(reviews_per_second)[0:4] \
                             + " Correct Predictions:" + str(correct_predictions) + " Tested:" + str(i+1) \
                             + " Testing Accuracy:" + str(correct_predictions * 100 / float(i+1))[:4] + "%")
    
    def run(self, review):
        self.input_layer *= 0
        indices = set()
        for word in review.lower().split(" "):
            if word in self.review_vocabulary_lookup_table.keys():
                indices.add(self.review_vocabulary_lookup_table[word])
        for index in indices:
            self.input_layer += self.weights_input_hidden[index]
        
        output_layer = self.sigmoid(self.input_layer.dot(self.weights_hidden_output))
    
        if(output_layer[0] >= 0.5):
            return "positive"
        else:
            return "negative"
```

```python
nn = NeuralNetwork(reviews[:-1000],labels[:-1000])
nn.train(reviews[:-1000],labels[:-1000])
```
